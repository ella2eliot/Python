```python
df = pd.read_csv('NLT_Topic.csv', encoding='utf-8')

df.head(3)
```

```python
current_year = 2024
df['加權: Weighted'] = 1 / (current_year - df['論文出版年: Thesis Publication Year'] + 1 )
df['加權點閱: Weighted Views'] = df['點閱: Views']*df['加權: Weighted']

df.to_csv('NLT_Topic_Weight.csv', index=False)
df
```

```python
# Creating a new column for category assignment based on the thesis title
def categorize_thesis(title):
    if "學習" in title or "教育" in title:
        return "Education and Learning Technology"
    elif "語言" in title or "詞彙" in title:
        return "Linguistics and Language Teaching"
    elif "遊戲" in title or "數位互動" in title:
        return "Digital interaction and game design"
    elif "平台" in title or "系統" in title:
        return "Technology application and platform construction"
    else:
        return "Interdisciplinary and theoretical research"

# Apply the categorization to the dataset
df['分類: Category'] = df['論文名稱: Thesis Title (Chinese)'].apply(categorize_thesis)

# Return the categorized dataframe
df.to_csv('NLT_Topic_Weight_with_Category.csv')
df.head(3)
```

```python
# 將同類的加權點閱相加
summed_views = df.groupby('分類: Category')['加權點閱: Weighted Views'].sum().reset_index()
summed_views.to_csv('Weighted views per category')
summed_views
```

```python
# 小數點整理
summed_views['加權點閱: Weighted Views'] = round(summed_views['加權點閱: Weighted Views'], 2)
```

```python
# Create a bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.bar(summed_views['分類: Category'], summed_views['加權點閱: Weighted Views'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Sum of Weighted Views')
plt.title('Sum of Weighted Views per Category')
# 顯示數值標籤
for i, v in enumerate(summed_views['加權點閱: Weighted Views']):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
