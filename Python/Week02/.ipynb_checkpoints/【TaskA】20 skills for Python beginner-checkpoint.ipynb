- For example
 - money = {'Tom' : 30000, 'Mary' : 64000, 'Gary' : 72000, 'Tim' : 500, 'Eva' : 6150}

- Please design a program to manage bank customers' money, customer names, and dates.
 - You must use input() and a dictionary (or list) to record each customers' money.
 - You should run and DEMO the function in our course's JupyterNB.
 - You need to create an interactive menu for the user, which should include at least the following functionalities:
  - 1. Allow the user to create multiple customers with their names, transaction times, and money at once.
  - 2. List all customers' money.
  - 3. User can deposit and withdraw money for customers.
  - 4. View transactions by date to see when an account deposited or withdrew money.

- 請設計一個程式管理銀行客戶的金錢、客戶名稱、日期。
 - 必須使用input()和字典(或list)來記錄客戶的金錢。
 - 需要為使用者編寫一個互動選單，其中至少包括以下功能：
 - 需要在本頁面一一執行你設計好的功能。
  - 1. 讓使用者一次創建多個客戶的名字、操作時間和金錢。
  - 2. 列出所有客戶的金錢。
  - 3. 使用者可以為客戶存錢及提款。
  - 4. 可透過日期查看，哪個帳戶在什麼時候存進/取出了多少錢。
  - 5. 至少一個上面沒提到的額外功能 例如：兌換外幣、轉帳、異常帳戶、月結算...等等。
  - 6. 只能使用list或dictionary，不能使用dataframe!!!!!!

```python
# 創建帳戶
def create_accounts():
    num_accounts = int(input('請輸入要創建的客戶數量(須為整數):'))
    for _ in range(num_accounts):
        name = str(input(f'請輸入第{_+1}位帳戶姓名:'))
        while name in customers:
            print('該帳戶已存在，請重新確認')
            name = str(input(f'請輸入第{_+1}位帳戶姓名:'))
        password = str(input('請為該帳戶創建密碼:'))
        time = input('請輸入目前操作時間(YYYYMMDD):')
        money = float(input('請輸入初始存入台幣金額(須為數字):'))
        customers[name] = {'date': time, 'pw': password, 'money': money, 'history': []}
        customers[name]['history'].append({'date': time, 'action': '創建', 'amount': money})
        print(f"帳戶 {name} 創建成功，目前有台幣 {money} 元")

# 驗證密碼
def password_verify(name, max_try):
    for i in range(1, max_try+1):
        password = str(input('請輸入帳戶密碼:'))
        if customers[name]['pw'] == password:
            return True
        else:
            j = 5 - i
            if j == 0:
                print('該帳戶已刪除')
                customers.pop(name)
                return False
            else:
                print(f'密碼輸入錯誤，還有{j}次可以嘗試')
        
# 存錢
def save_money():
    name = str(input('請輸入要存入的帳戶姓名:'))
    if name in customers:
        if password_verify(name, 5):
            add_money = float(input('請輸入要存入的台幣金額:'))
            time = input('請輸入目前操作時間(YYYYMMDD):')
            customers[name]['money'] += add_money
            money = customers[name]['money']
            customers[name]['history'].append({'date': time, 'action': '存錢', 'amount': add_money})
            print(f'台幣 {add_money} 元已存入 {name} 的帳戶，目前有 {money} 元') # 在'前面使用f可以保留{}內的變數而不將其視為str
    else:
        print('找不到此客戶')
        
# 提款/轉帳
def withdrawal():
    name = str(input('請輸入要提款/轉帳的帳戶姓名:'))
    if name in customers:
        if password_verify(name, 5):
            withdraw_or_transfer = input('請問要提款還是轉帳(A為提款B為轉帳，請輸入A/B):')
            if withdraw_or_transfer == 'A':
                withdraw_money = float(input('請輸入要提款的台幣金額:'))
                time = input('請輸入目前操作時間(YYYYMMDD):')
                customers[name]['money'] -= withdraw_money
                money = customers[name]['money']
                customers[name]['history'].append({'date': time, 'action': '提款', 'amount': withdraw_money})
                print(f'台幣{withdraw_money} 已從 {name} 的帳戶取出，目前剩餘 {money} 元') # 在'前面使用f可以保留{}內的變數而不將其視為str
            else:
                while True:
                    transfer_name = str(input('請輸入要轉帳的對象:'))
                    if transfer_name in customers:
                        transfer_money = float(input('請輸入要轉帳的台幣金額:'))
                        time = input('請輸入目前操作時間(YYYYMMDD):')

                        # 檢查餘額是否足夠
                        if customers[name]['money'] >= transfer_money:
                            customers[name]['money'] -= transfer_money
                            customers[transfer_name]['money'] += transfer_money
                            withdraw_money = transfer_money

                            # 記錄轉帳歷史
                            customers[name]['history'].append({'date': time, 'action': '轉帳', 'amount': transfer_money})
                            customers[transfer_name]['history'].append({'date': time, 'action': '入帳', 'amount': transfer_money})
                            print(f"台幣 {transfer_money} 元已從 {name} 的帳戶轉至 {transfer_name}，目前剩餘 {customers[name]['money']} 元")
                            print(f"{transfer_name} 的金額目前為 {customers[transfer_name]['money']} 元")
                            break
                        else:
                            print('餘額不足，無法完成轉帳')
                            break
                    else:
                        print('該帳戶不存在，請重新輸入轉帳對象')
    else:
        print('找不到此客戶')
        
# 列出所有客戶金額
def money_list():
    if not customers:
        print("目前沒有客戶資料")
    else:
        for name, details in customers.items():
            print(f"客戶: {name}, 台幣金額: {details['money']}")

# 日期查詢
def date_check():
    name = str(input('請輸入要查詢的帳戶姓名:'))
    if name in customers:
        if password_verify(name, 5):
            date = input('請輸入要查詢的日期(YYYYMMDD):')
            history = customers[name]["history"]
            check = False
            for transaction in history: # transaction在此表示customers[name]["history"][01234...]的意思
                if transaction['date'] == date:
                    print(f"日期: {date}, 操作: {transaction['action']}, 金額: 台幣 {transaction['amount']} 元")
                    check = True
            if not check:
                print('在該日期並無任何操作')
    else:
        print('找不到此客戶')

# 註銷帳戶
def del_account():
    name = input('請輸入要註銷的帳戶姓名:')
    if name in customers:
        if password_verify(name, 5):
            confirm_del = input('你確定要註銷帳戶嗎?(Y/N)')
            if confirm_del == 'Y':
                confirm_del1 = input('你真的要刪除嗎?(Y/N)')
                if confirm_del1 == 'Y':
                    confirm_del2 = input('你發誓註銷之後不會後悔嗎?(Y/N)')
                    if confirm_del2 == 'Y':
                        print('該帳戶已刪除')
                        customers.pop(name)
                    else:
                        print('非常感謝您願意拯救我的業績~')
                else:
                    print('感謝您的體恤~')
            else:
                print('謝謝您~')
    else:
        print('該帳戶不存在')
```

```python
# 定義互動選單
def menu():
    global customers
    # 客戶資料
    customers = {}
    while True:
        print("\n1. 創建客戶")
        print("2. 列出所有客戶的金錢")
        print("3. 存款")
        print("4. 提款/轉帳")
        print("5. 根據日期查看交易記錄")
        print("6. 註銷帳戶")
        print("7. 退出")
        choice = input("請選擇操作:")

        if choice == "1":
            create_accounts()
        elif choice == "2":
            money_list()
        elif choice == "3":
            save_money()
        elif choice == "4":
            withdrawal()
        elif choice == "5":
            date_check()
        elif choice == "6":
            del_account()
        elif choice == "7":
            print("退出系統")
            break
        else:
            print("無效的選擇，請重試")
```

```python
menu()
```
